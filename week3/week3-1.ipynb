```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Data-cleaning-for-beginners-using-pandas.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
avg_age = df.Age.mean()
avg_age
```

```python
df.Age = df.Age.fillna(avg_age)
df.Age = df.Age.round(decimals=1)
df.head()
```

```python
df.isnull().sum()
```

```python
avg_rtg = df.Rating.mean()
df.Rating = df.Rating.fillna(avg_rtg)
df.Rating = df.Rating.round(decimals=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
df.head()
```

```python
df.established = df.established.replace(-1,2000)
df.head()
```

```python
df.rating = df.rating.replace(-1.0, 0)
df.head()
```

```python
for i in df.salary:
    j = i.replace("$","")
    j = j.replace("k","000")
    df.salary = df.salary.replace(i, j)
df.head()
```

```python
df.info()
```

```python
df.easy_apply = df.easy_apply.replace(-1, 0)
df.head()
```

```python
df.easy_apply = df.easy_apply.replace("TRUE", 1)
df.head()
```

```python
df.easy_apply = df.easy_apply.astype('int64')
```
